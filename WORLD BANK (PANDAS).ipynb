### Plotting series using pandas

```python
# Austin temperature data
temperature_data = 'C://Users//paul//Documents//STUTERN PANDAS PROJECT//2010 Austin weather.csv'

# read in the data set
austin = pd.read_csv(temperature_data)
```

```python
austin
```

```python
# temperature column 
temp = austin.iloc[:700, 0]

# Create a plot with color='red'
temp.plot(color = 'red')

# title
plt.title('Temperature in Austin')

# Specify the x-axis
plt.xlabel('Hours since midnight August 1, 2010')

# Specify the y-axis label
plt.ylabel('Temperature (degrees F)')

# Display the plot
plt.show()
```

```python
#  Transposing the messy data we cleaned in order to get the stock prices data
df_e = df_d.transpose()
df_e
```

## Scatter plots

```python
# import the dataset for the task
auto_mobile = pd.read_csv('C://Users//paul//Documents//STUTERN PANDAS PROJECT//Automobile miles per gallon.csv')

auto_mobile
```

```python
# numpy array of car weight 
weight = np.array(auto_mobile['weight'])

# mean of the weights
mean_weight = weight.mean()

sizes = weight/mean_weight # get the normal form of an array, Divide an array by its mean
sizes
```

```python
# scatter plot
auto_mobile.plot(kind='scatter', x='hp', y='mpg', s=sizes)

# title
plt.title('Fuel efficiency vs Horse-power')

#  x-axis label
plt.xlabel('Horse-power')

# y-axis label
plt.ylabel('Fuel efficiency (mpg)')

plt.show()
```

```python
# Make a list of the column names to be plotted: cols
cols = ['weight', 'mpg']


auto_mobile[cols].plot(kind='box', subplots=True, figsize = (8,4))

plt.show()
```
